# **Menghapus Row Missing Value (NaN) Pada Files Order.csv**

```python
order = order.dropna().drop('created_at', 1).drop('shipped_at', 1).drop('delivered_at', 1).drop('returned_at', 1)
order
```

```python
items = items.dropna()
items
```

# **Menggabungkan Files Order.csv dengan Order Items.csv tanpa Missing Value (NaN)**

```python
join_order = pd.merge(items, order, how='left', on=['order_id','user_id','status'])
join_order
```

# **Mengubah Files Join_Order.csv Kedalam SQL Dan Menampilkan Data Join_Order.sql**

Variabel List
- id	
- order_id	
- user_id	
- product_id	
- inventory_item_id	
- status	
- created_at	
- shipped_at	
- delivered_at	
- returned_at	
- sale_price	
- gender	
- num_of_item

```python
import sqlite3
import csv

conn = sqlite3.connect('/content/drive/MyDrive/Colab Notebooks/sales/join_order.sqlite')
cur = conn.cursor()

cur.execute('DROP TABLE IF EXISTS join_order')
cur.execute('''
    CREATE TABLE "join_order"(
      "id" TEXT,
      "order_id" TEXT,
      "user_id" TEXT,
      "product_id" TEXT,
      "inventory_item_id" TEXT,
      "status" TEXT,
      "created_at" TEXT,
      "shipped_at" TEXT,
      "delivered_at" TEXT,
      "returned_at" TEXT,
      "sale_price" REAL,
      "gender" TEXT,
      "num_of_item" INT
    )
''')

fname = '/content/drive/MyDrive/Colab Notebooks/sales/join_order.csv'
if len(fname) < 1 : fname = fname

with open(fname) as csv_file:
  csv_reader = csv.reader(csv_file, delimiter=',')
  for row in csv_reader:
    print(row)
    id=row[0]
    order_id=row[1]
    user_id=row[2]
    product_id=row[3]
    inventory_item_id=row[4]
    status=row[5]
    created_at=row[6]
    shipped_at=row[7]
    delivered_at=row[8]
    returned_at=row[9]
    sale_price=row[10]
    gender=row[11]
    num_of_item=row[12]
    cur.execute(
        '''INSERT INTO join_order(id,order_id,user_id,product_id,inventory_item_id,status,created_at,shipped_at,delivered_at,returned_at,sale_price,gender,num_of_item) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)''',
        (id,order_id,user_id,product_id,inventory_item_id,status,created_at,shipped_at,delivered_at,returned_at,sale_price,gender,num_of_item)
    )
    conn.commit()
```
