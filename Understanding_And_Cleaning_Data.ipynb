# <img src="https://img.icons8.com/color/48/undefined/3-circle--v1.png"/>**Menampilkan Data Order Items**

```python
items = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/sales/order_items.xlsx")
items
```

# <img src="https://img.icons8.com/color/48/undefined/4-circle--v1.png"/>**Menghapus Row Missing Value (NaN) Pada Files Order.excel**

```python
order = order.dropna().drop('created_at', 1).drop('shipped_at', 1).drop('delivered_at', 1).drop('returned_at', 1)
order
```

# <img src="https://img.icons8.com/color/48/undefined/5-circle--v1.png"/>**Menghapus Row Missing Value (NaN) Pada Files Order Items.excel**

```python
items = items.dropna()
items
```

# <img src="https://img.icons8.com/color/48/undefined/6-circle--v1.png"/>**Menggabungkan Files Order.excel dengan Order Items.excel tanpa Missing Value (NaN)**

```python
join_order = pd.merge(items, order, how='left', on=['order_id','user_id','status'])
join_order
```

# <img src="https://img.icons8.com/color/48/undefined/7-circle--v1.png"/>**Menyimpan Hasil Join Files Order.excel dan Order items.excel dengan nama join_order.excel**

```python
join_order.to_excel("/content/drive/MyDrive/Colab Notebooks/sales/join_order.xlsx", index=False)
```

# <img src="https://img.icons8.com/color/48/undefined/8-circle--v1.png"/>**Mengubah Files Join_Order.excel Kedalam SQL Dan Menampilkan Data Join_Order.sql**

Variabel List

`id`,	`order_id`, `user_id`, `product_id`, `inventory_item_id`, `status`, `created_at`, `shipped_at`, `delivered_at`, `returned_at`, `sale_price`, `gender`, `num_of_item`

```python
import sqlite3
import pandas as pd
filename="/content/drive/MyDrive/Colab Notebooks/sales/join_order"
con=sqlite3.connect(filename+".db")
wb=pd.ExcelFile(filename+'.xlsx')
for sheet in wb.sheet_names:
        df=pd.read_excel(filename+'.xlsx')
        df.to_sql(sheet,con, index=False,if_exists="replace")
con.commit()
con.close()
```
